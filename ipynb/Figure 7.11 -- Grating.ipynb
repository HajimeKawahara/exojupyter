```python
import pylab
import matplotlib.pyplot as plt
import numpy as np
```

```python
def Hfunc(N,x):
    return np.sin(np.pi*N*x)**2/np.sin(np.pi*x)**2
```

```python
def sinc(x):
    return np.sin(np.pi*x)**2/(np.pi*x)**2
```

```python
xlist=np.linspace(-10,10,10000)
fig=plt.figure()
ax=fig.add_subplot(211)
pylab.xlabel("$X$")
pylab.ylabel("$H_N(X)$")
pylab.ylim(0,110)
pylab.xlim(-10,10)
ax.plot(xlist,Hfunc(10,xlist),label="$N=10$")
ax.plot(xlist,Hfunc(5,xlist),lw=2,label="$N=5$")
ax.legend()
ax=fig.add_subplot(212)
pylab.xlabel("$X$")
pylab.ylabel("$H_N(X) |\\tilde{M}|^2$")
pylab.ylim(1,110)
pylab.xlim(-10,10)
ax.plot(xlist,Hfunc(10,xlist)*sinc(xlist*0.2),label="$N=10$")
ax.plot(xlist,Hfunc(5,xlist)*sinc(xlist*0.2),lw=2,label="$N=5$")
ax.legend()
plt.show()
```

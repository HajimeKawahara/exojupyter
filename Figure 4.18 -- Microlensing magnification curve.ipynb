```python
import pylab
import matplotlib.pyplot as plt
import numpy as np
```

```python
def mut(beta):
    muA=(beta**2+2.0)/(beta*np.sqrt(beta**2+4))
    return muA
```

```python
def betat(tau,u0):
    #tau = (t-t0)/tE
    #u0=beta/thetaE at the closest
    beta = np.sqrt(u0**2+tau**2)
    return beta
```

```python
taul=np.linspace(-2,2,1000)
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(taul,mut(betat(taul,0.1)),ls="solid")
ax.plot(taul,mut(betat(taul,0.3)),ls="dashed")
ax.plot(taul,mut(betat(taul,0.7)),ls="dotted")
plt.legend(["$\\beta_0 = 0.1 \\theta_E$","$\\beta_0 = 0.3 \\theta_E$","$\\beta_0 = 0.7 \\theta_E$"],loc="upper right",prop={'size':11},frameon=False)
#ax.legend(["$\beta_0 = 0.1 \theta_E$","$\beta_0 = 0.3 \theta_E$","$\beta_0 = 0.7 \theta_E$"])
plt.tick_params(labelsize=14)
plt.ylim(0,12)
plt.ylabel("magnification $\mu_A$",fontsize=14)
plt.xlabel("normalized time $(t-t_0)/t_E$",fontsize=14)
#plt.savefig("pointmag.pdf", bbox_inches="tight", pad_inches=0.0)
plt.show()
```
